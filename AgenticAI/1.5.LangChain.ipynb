## what is langchain ? 
it is a frameworks for developing application powered by LLMs . 
Instead of writing raw API calls and handling everything manually, LangChain gives you ready-made building blocks to create smarter AI systems.
it simplies the app development process . 

## benefits of langchain : 

1. concept of chains : we can connect different componets using a chain or like a pipeline . 
the output of one component will become input of another . 
we can create sequental , paraller or conditional , complex chains . 

2. model agnostic development : we can build system that does not depend on a specific model . like we can change the model by using only one line . 

3. Complete ecosystem : langchain tries to provide a ecosystem of any component used in the chain , like we can choose any from ecosystem for a specific component . 

4. memory and state handling .
