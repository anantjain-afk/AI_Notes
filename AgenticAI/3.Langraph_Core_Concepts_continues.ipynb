Node : represents a single task . 
every node is a python function behind the scene. 
edges : represents what next task will execute after a perticular task is done being executed .
so edges connect the nodes .
graph is just for ease of expressing the system . 



State : 
it is a piece of information or shared memory which is necessary for the execution to continue . and this state evolves overtime . it holds all data beign passed between nodes as graph runs .
we need to define state before running the graph . 
it is just a key pair dictionry . (TypedDict) in python 
at any moment every node have access to this state .



Reducers : 
it defines how updates from nodes are applied to the shared state .
each key in the state can have its own reducer , which determines wheather new data replaces , merge or add to its existing value . 


Execution model  : 
1. Graph Definition : 
State Schema , nodes , edges . 

2. compilation : 
this checks the graph structure and prepares it for execution .
you call .combine() on the StateGraph . 

3. Invocation : 
you run the graph with .invoke(initial_state)
initial_state is the first node or root node . it will run the function . and does some partial changes to the state . 

4. Super - step begins : 
execution process in rounds . 

5. Message passing & Node activation :
The messages are passed to downstream nodes via edges
Nodes that receive messages become active for the next round

6. Halting condition:
Execution stops when
- No nodes are active, and
- No messages are in transit
