## LangGraph 
### why langGraph exists ? 



### challenges in langchain : 
1. Control flow complexity : it is not good for non linear flow . 
a. Conditional flow 
b. loops 
c. jumps 
d. glue code 

whereas in graph you have nodes . and edges decides the flow , we can represent any non linear flow easily and we can code it in a simple way. 

2. State Handling : 
Keeping track of all information as it flows through different steps (nodes) of your AI workflow
in langGraph we have state varaible , it is very easy to handle state . 




in langchain there is no way to store this key value pair of current state , it has to be done manually in langchain . 
whereas in langgraph we have stateful workflow. where at any time , we have all state available at all nodes .

3. Challenge 3 :  Event Driven Execution : 
event driven execution is that workflow whcih need some type of action to be done first . 
but langchain is not built for event driven execution its made for sequential like once it start it end after doing the complete the chain , it will not wait for any event . 
whereas langgraph provide inherit event driven executio .
we can create checkpoints in langgraph . where the agent can wait for external events to be done . 


4. Fault Tolerance : system is working even after errors . 
Fault : a.  small (as in node level) 
b. big fault (sever crashed ) .
in langchain -> as the workflow is a chain if one component fails the work will start from begingin . 
whereas langgraph we have retry logic for small fault , like if something crashed , we can wait and try again after some time . it generally solves the small fault. 
for big fault : langgraph have recovery concept .
if let's say sever get's crashed at some node, we can continue from that node only as we already have checkpoint created for it .


5. Human in the loop : 
in langchain , it is a big problem as we don't have any default feature by which the sequentail syncronous chain can pause for some time to ask for human  review , although it can wait for small time for human input , but its not made for long wait. 

in langgraph : 
it has built in feature for it using statefulness and checkpoint . 


6.  Nested workflow : 
in langgraph it is possible to repplace a node with another graph or subgraph hence we create nested worflow . 
we can create multi agent system . 
we can also create re-graph and use it anywhere so its reusable component . 

7. Observability :
it refers to how easify you can monitor , debug and understand what your workflow  is doing at runtime .
its also in langcahin using langsmith . 
but it is partial observability it can not track glue code .
where langgraph have timelines , it can log everything . 






### what is Langgraph :
it is a orchestrator framework which allows you to create stateful , event driven , multi-step workflows using LLMs , it is ideal for designing single agent and multi - agent agentic ai workflows. 
it has steps (nodes) , we can see how they are connected and governs how transitions happens . langgraph takes cares of conditional branching , looping , pausing and fault recovery . 

### langchain vs langgraph ?? when to use what ? 
use langchain when to create simple sequential workflows - like promt chain , summarizer  
use Langgraph when your use case involves complex non linear workflows that needs conditional branchign , looping , fault taulrance , human in the loops ,, etc .
