Tokenization refers to break down the text into smaller units. It splits paragraphs into sentences and sentences into words. It is one of the initial steps of any NLP pipeline.
Reduces the size of raw text, resulting in easy and efficient statistical and computational analysis


```python

import nltk
# downloads auxiliary statistical tables required by the Punkt tokenizer to correctly identify sentence boundaries, especially for abbreviations and punctuation.
nltk.download('punkt_tab')
```

```python
from nltk import sent_tokenize, word_tokenize
sent = "GeeksforGeeks is a great learning platform . It is one of the best for Computer Science students."
print(word_tokenize(sent))
print(sent_tokenize(sent))
```
