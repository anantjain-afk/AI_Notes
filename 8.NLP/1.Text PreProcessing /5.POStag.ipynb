### Parts of Speech (POS) is a fundamental task in nlp where each word is assigned its corresponding relevant tag like verb , noun , etc . 
#### we do this to help machine understands the meaning and structure of sentences by identifying the roles of words in the sentence . 


```python
import nltk
from nltk import word_tokenize , pos_tag
# we have to download the averaged_perceptron_tagger_eng for pos tagging. 
nltk.download('averaged_perceptron_tagger_eng')
sent = "The striped bats are hanging on their feet for best"
words = word_tokenize(sent)
pos_tags = pos_tag(words)
for word, pos_tag in pos_tags:
    print(f"{word}: {pos_tag}")
```
