Lemmatization is like stemming , the output that we get is called "lemma" here , which is a root word rather than root stem . so , after lemmatization we will get valid word which will have some meaning unlike root stem in stemming . NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus(dictionary) . This class uses morphy() function to the WordNet CorpusReader class to find a lemma. 


```python
import nltk
from nltk import WordNetLemmatizer 
# we can use pre defined dectionary provided by nltk to lemmatize words ,
#  this dectionary contains root or base words and their various forms
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
'''POS- 
Noun-n 
verb-v 
adjective-a 
adverb-r'''
# POS => part of speech . It is used to specify the type of word we are lemmatizing
# like noun , verb , adjective , adverb etc
# default pos is noun 
lemmatizer.lemmatize("going") #going
print(lemmatizer.lemmatize("going", pos='v')) #go
# see the difference when we change the pos
lemmatizer.lemmatize("better", pos='a') #good
```

 To improve the accuracy of lemmatization, it’s important to specify the correct Part
 of Speech (POS) for each word. By default, NLTK assumes that words are nouns when no POS tag is provided. However, it can be more accurate if we specify the correct POS tag for each word.

```python
words=["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "programmed" , "Historys"]

for word in words:
    print(f"{word} : {lemmatizer.lemmatize(word, pos='v')}")
```

Lemmatization takes more time compared to stemming . 
Due to its time-consuming nature, it may not be well-suited for real-time applications where fast processing is important.
