## Bag of Words Model 
In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the words or their grammar but focuses on counting how often each word appears in the text.

## Step-by-Step: After Text Preprocessing
rawStet = "The cats are running fast!" 
after processing : ["cat", "run", "fast"]
### STEP 1: Build Vocabulary 
Collect unique words from the entire corpus   
Example corpus : [Doc1: "cat run fast",Doc2: "dog run slow"].  
Vocabulary : ["cat", "run", "fast", "dog", "slow"]. 
### WHY we do this

1.ML models need fixed dimensions

2.Vocabulary defines feature space

3.Each word becomes one feature

### STEP 2: Assign Index to Each Word 
cat  → 0
run  → 1
fast → 2
dog  → 3
slow → 4
### WHY we do this

Computers need positions, not words

Index = column number in matrix

### STEP 3: Create Vectors for Each Document
Count word occurrences in each document.   
doc1 = "cat run fast" , 
Vector = [1, 1, 1, 0, 0]
doc2 =  "dog run slow" , Vector : [0, 1, 0, 1, 1]

### Step 4 : Form Document Matrix : 
Why : Standard input format for ML models . 

### Advanges:
1. Easy and intuitive . 
2. Fixed Dimention of Ml model I/P . good for ml algo. 


### Same types of limitation are here too : 
1. Overfitting 
2. No semantic 
3. Sparse matrix
4. No context or order . 
5. Out of Vocab => new word appears in the sentence 


## Implementing Bag of Words with Scikit-learn(sklearn)

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer 
import numpy as np
# Step 1: Preparing the Corpus
corpus = [
    "The cat sat on the mat.",
    "The dog sat on the log."
]
# Step 2: Initializing the Bag of Words Model
vectorizer = CountVectorizer() #we can also provide parameters like stop_words, max_features, binary etc.
# for binary bow : CountVectorizer(binary=True)
# Step 3: Fitting and Transforming the Corpus
X = vectorizer.fit_transform(corpus)
# fit_transform will do 
# 1. text preprocessing (like tokenization, lowercasing, removing punctuation)
# 2. building the vocabulary 
# 3. encoding the documents into vectors (Shape = (number_of_docs, vocabulary_size))

#Step 4 : displaying the vocabulary and frequency count. 
print("Vocabulary:", vectorizer.vocabulary_)
print("Features Names:", vectorizer.get_feature_names_out())
print("Encoded Vectors:\n", X.toarray())

#step 5:Analyzing Word Frequencies
word_counts = np.sum(X.toarray(), axis=0)
word_freq = dict(zip(vectorizer.get_feature_names_out(), word_counts))
print("Word Frequencies:", word_freq)


```
