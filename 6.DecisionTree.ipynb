A Decision Tree helps us to make decisions by mapping out different choices and their possible outcomes. It’s used in machine learning for tasks like classification and prediction.
A Decision Tree is a machine learning algorithm that builds a Tree to make decisions. It takes a dataset and splits it up by asking a series of questions.
leaf node is the final answer or decision that the algoritm makes . 



Decision Trees are needed because they provide interpretable, non-linear decision boundaries, handle categorical features naturally, and require minimal preprocessing, making them highly practical for real-world business applications.

## Impurity Measure (Gini Impurity)

Impurity : this is the messiness in the data . if the data has only one category then the data is completely pure . 

Gini Impurity is a mathematical score that measures this messiness. The score ranges from 0 (perfectly pure) to 0.5 (completely mixed up, assuming a two-category problem) for binary classification means only two categories exists. 
The tree calculates Gini using the probability (pi) of randomly picking an item of a specific class ($i$) from the group

Gini = 1- sum(pi)**2 where i is the class or category . 
The tree looks at all possible questions it could ask, calculates the Gini Impurity for the groups that question would create, and tries to find the lowest possible Gini score




## Information Gain 

if the gini is the messiness , information gain measures how much messiness we have removed . 

Think of it as a simple subtraction problem:
Impurity Before the Question - Impurity After the Question = Information Gain

If a question doesn't separate the data well at all, the impurity stays high, and the Information Gain is low.

If a question perfectly separates cats from dogs, the impurity drops to zero, and the Information Gain is massive.

At every single step, the Decision Tree algorithm chooses the question that provides the highest Information Gain.
Information Gain = entropy ( parent) — [average entropy ( children)]


## limitations :

Prone to overthinking : If you don't force them to stop growing, they will keep asking questions until they have just memorized the training data, making them terrible at predicting new data. 

Unstable: A tiny change in your training data can cause the algorithm to draw a completely different tree.

## regression Tree : 

Up until now, we've talked about predicting categories (like "Cat" vs "Dog" or "Spam" vs "Not Spam"). This is called Classification.

But what if you want to predict a specific, continuous number—like the price of a house or the temperature tomorrow? That is where Regression Trees come in. The logic is the same, but the math changes slightly.

No Categories: Instead of sorting items into groups like "Red" or "Blue," it sorts items into buckets of numbers.

No Gini: You can't calculate Gini Impurity on continuous numbers. Instead, Regression Trees measure Variance (how spread out the numbers are in a bucket). It tries to ask questions that group similar numbers together (low variance).

The Final Answer: When you reach a Leaf Node at the bottom of a Regression Tree, the prediction is simply the average of all the numbers that landed in that specific bucket.
